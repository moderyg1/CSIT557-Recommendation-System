```python
from surprise import Dataset, Reader, KNNBasic, accuracy
import pandas as pd
from surprise.model_selection import train_test_split
```

```python
columns = ["userId", "movieId", "categoryId", "reviewId", "movieRating", "reviewDate"]
df = pd.read_csv("../data/movie-ratings.txt", header=None, names=columns)

reader = Reader(rating_scale=(1, 5))

data = Dataset.load_from_df(df[['userId', 'movieId', 'movieRating']], reader)
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)
```

```python
sim_options_cosine = {
    'name': 'cosine',
    'user_based': False
}

model_cosine = KNNBasic(sim_options=sim_options_cosine)

# Train
model_cosine.fit(trainset)

# Predict
predictions_cosine = model_cosine.test(testset)

# Evaluate
mae_cosine = accuracy.mae(predictions_cosine)
rmse_cosine = accuracy.rmse(predictions_cosine)
```

```python
sim_options_pearson = {
    'name': 'pearson',
    'user_based': False
}

model_pearson = KNNBasic(sim_options=sim_options_pearson)

# Train
model_pearson.fit(trainset)

# Predict
predictions_pearson = model_pearson.test(testset)

# Evaluate
mae_pearson = accuracy.mae(predictions_pearson)
rmse_pearson = accuracy.rmse(predictions_pearson)
```

```python
print("Cosine Similarity:")
print(f"MAE: {mae_cosine}")
print(f"RMSE: {rmse_cosine}")

print("\nPearson Similarity:")
print(f"MAE: {mae_pearson}")
print(f"RMSE: {rmse_pearson}")
```

```python
pd.DataFrame({
    "MAE": [mae_cosine, mae_pearson],
    "RMSE": [rmse_cosine, rmse_pearson]
}, index=["Cosine", "Pearson"]).to_csv("../results/ItemBasedCF_Results.csv")
```
